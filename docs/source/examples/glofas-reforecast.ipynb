# GloFAS reforecast

```python
import climetlab as cml
import climetlab_cems_flood as cmf
```

```python
nwse = [50.972204,5.450796, 46.296530, 11.871059] # Ryne
```

```python
ref = cml.load_dataset(
            'cems-flood-glofas-reforecast',
            model='lisflood',
            product_type='ensemble_perturbed_reforecasts',
            system_version='version_3_1',
            period= '20180710-20180720',
            leadtime_hour = '24-72',
            variable="river_discharge_in_the_last_24_hours",
            area= nwse,
            split_on = ['hday'],
            threads = 6,
            merger=None
        )
```

```python
ref.to_netcdf("/data/temp","new")
```
